```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import active_contour
from skimage.color import rgb2gray
from skimage import img_as_float

# =========================
# LOAD GAMBAR
# =========================
img = cv2.imread('tomat.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize agar stabil
img_rgb = cv2.resize(img_rgb, (400, 400))

# =========================
# PREPROCESSING
# =========================
gray = rgb2gray(img_rgb)
gray = img_as_float(gray)

# Optional: blur biar lebih halus
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)

# =========================
# INISIALISASI CONTOUR
# =========================
h, w = gray.shape

s = np.linspace(0, 2*np.pi, 300)
r = min(h, w) // 3   # radius otomatis

x = w//2 + r * np.cos(s)
y = h//2 + r * np.sin(s)
init = np.array([x, y]).T

# =========================
# ACTIVE CONTOUR
# =========================
snake = active_contour(
    gray_blur,
    init,
    alpha=0.01,   # elastisitas
    beta=10,      # kekakuan
    gamma=0.001
)

# =========================
# TAMPILKAN 4 OUTPUT
# =========================
fig, ax = plt.subplots(1, 4, figsize=(16,4))

# 1. Original
ax[0].imshow(img_rgb)
ax[0].set_title("Original")
ax[0].axis('off')

# 2. Grayscale
ax[1].imshow(gray, cmap='gray')
ax[1].set_title("Grayscale")
ax[1].axis('off')

# 3. Initial Contour
ax[2].imshow(gray, cmap='gray')
ax[2].plot(init[:, 0], init[:, 1], '--r', lw=2)
ax[2].set_title("Initial Contour")
ax[2].axis('off')

# 4. Hasil Active Contour
ax[3].imshow(gray, cmap='gray')
ax[3].plot(snake[:, 0], snake[:, 1], '-b', lw=2)
ax[3].set_title("Active Contour Result")
ax[3].axis('off')

plt.tight_layout()
plt.show()
```

```
# Convert ke HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Range warna (contoh: merah)
lower = np.array([0, 120, 70])
upper = np.array([10, 255, 255])

mask = cv2.inRange(hsv, lower, upper)
result = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(mask, cmap='gray')
plt.title("Mask Warna")

plt.subplot(1,2,2)
plt.imshow(result)
plt.title("Segmentasi Warna")
plt.axis('off')
```
